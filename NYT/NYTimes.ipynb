```python
import csv
import urllib.request
import os
from bs4 import BeautifulSoup

def writeArticle(csvname,topic):
    with open(csvname) as csvfile:
        #csvfile="NYT-Movie.csv"
        readCSV = csv.reader(csvfile)
        count=0
        for row in readCSV:
            url=row[0] #'https://www.nytimes.com/2019/01/11/movies/studio-ghibli-ponoc-anime.html'
            try:
                content = urllib.request.urlopen(url).read()
            except:
                continue
            soup = BeautifulSoup(content, "html.parser")
            # Opening the file
            count=count+1
            fname='./NYTarticles/'+'article'+topic+str(count)+'.txt'
            file = open(fname, "a+")

            # Getting the article body and saving to the file
            for a in soup.find_all('p'):
                file.write(a.text)
                file.flush()
                #print(a.text)
            file.close()
            if count == 150:
                print("breaking")
                break
            
                        
writeArticle('./NYT-url/NYT-TV1.csv','TvNYT')            
```
