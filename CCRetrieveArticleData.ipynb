```python
import csv
import urllib.request
import os
from bs4 import BeautifulSoup
import pandas as pd
import os
import requests
```

```python
def writeArticle(csvname,topic):
    with open(csvname) as csvfile:
        #csvfile="NYT-Movie.csv"
        readCSV = csv.reader(csvfile)
        count=0
        for row in readCSV:
            url=row[0] #'https://www.nytimes.com/2019/01/11/movies/studio-ghibli-ponoc-anime.html'
            
            #print (url)
            try:
                content = requests.get(url).content
            except:
                continue
            soup = BeautifulSoup(content, "html.parser")
            # Opening the file
            count=count+1
            fname='./CommonCrawlData/'+'Article'+topic+str(count)+'.txt'
            file = open(fname, "w+")
            print("Writing "+str(count) + " : "+url)
            # Getting the article body and saving to the file
            for a in soup.find_all('p'):
                file.write(a.text)
                file.flush()
#                 print(a.text)
            for a in soup.find_all('div'):
                file.write(a.text)
                file.flush()
#                 print(a.text)
#                 print("----Next----")
            file.close()
            if count == 120:
                break
        return count
            
            
#writeArticle('NYT-.csv','Trial')
```

```python
CCLinksLoc = "./CommonCrawlData/CC-Links/"
CCLinks = os.listdir(CCLinksLoc)
print (CCLinks)
for link in CCLinks:
    if "csv" in link and  "Tele" not in link and "Game" not in link and "Music" not in link:
        
        topic = link.split("-")[1]
        print (topic, link)
        countedWrites = writeArticle(CCLinksLoc + link, topic)
        print(topic + " written is "+ str(countedWrites))
```
